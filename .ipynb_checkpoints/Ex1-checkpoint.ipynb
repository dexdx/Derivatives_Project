```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RCN_terminal(alpha, i_0, i_T):
    strike = alpha*i_0
    if i_T > strike:
        return 1
    else:
        return 1 - alpha + i_T/i_0
```

```python
alpha = 1
i_0 = 10
i_T = np.arange(0,20,.01)
RCN_terminal_vec = np.vectorize(RCN_terminal)
H_T = RCN_terminal_vec(alpha, i_0, i_T)
```

```python
plt.plot(i_T, H_T)
plt.xlabel("$I_T$", fontsize=16)
plt.ylabel("$H(I_T)$", fontsize=16);
```
